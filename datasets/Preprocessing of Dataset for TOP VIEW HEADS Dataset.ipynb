# Preprocessing:
 + **Step# 1:** First I have separated the 8 bit images, 16 bit images and masks from the Original Dataset to separate folders.
 + **Step# 2:** Load and read the images and respective masks and save to Numpy Arrays.



```
import os
import shutil
import numpy as np
from tqdm import tqdm_notebook as tqdm
from keras.preprocessing.image import load_img, img_to_array

#Step# 1:
os.chdir('/content/drive/My Drive/GitHub Repositories/State of Art Segmentation Network Architectures')
dataDir = './datasets/TOP-VIEW HEADS DATASET'
imgNames = sorted(os.listdir(dataDir))
print('Number of Images: {}'.format(len(imgNames)))

images8bit_Dir = './datasets/TOP-VIEW HEADS DATASET/Images(8 Bit)'
images16bit_Dir = './datasets/TOP-VIEW HEADS DATASET/Images(16 Bit)'
masksDir = './datasets/TOP-VIEW HEADS DATASET/Masks'

if not os.path.exists(images8bit_Dir):
    os.makedirs(images8bit_Dir)
    
if not os.path.exists(images16bit_Dir):
    os.makedirs(images16bit_Dir)

if not os.path.exists(masksDir):
    os.makedirs(masksDir)

for imgName in tqdm(imgNames):
    temp_imgName = imgName.rsplit('.', 1)[0]
    temp_imgName = temp_imgName.rsplit('_', 1)[1]
    if temp_imgName == '8':
        shutil.copy('./datasets/TOP-VIEW HEADS DATASET/Original dataset/'+str(imgName), images8bit_Dir+'/'+str(imgName))
    elif temp_imgName == 'mask':
        shutil.copy('./datasets/TOP-VIEW HEADS DATASET/Original dataset/'+str(imgName), masksDir+'/'+str(imgName))
    else:
        shutil.copy('./datasets/TOP-VIEW HEADS DATASET/Original dataset/'+str(imgName), images16bit_Dir+'/'+str(imgName))

#Step# 2:
imagesDir = './datasets/TOP-VIEW HEADS DATASET/Images(8 Bit)'
masksDir = './datasets/TOP-VIEW HEADS DATASET/Masks'

imageNames = sorted(os.listdir(imagesDir))
maskNames = sorted(os.listdir(masksDir))

images = []
masks = []

img_rows = 96
img_cols = 128
channels = 1

for imageName in tqdm(imageNames):
  temp = imageName.rsplit('_', 1)[0]
  maskName = temp + '_mask.png'
  if maskName in maskNames:
    img = load_img(os.path.join(imagesDir, imageName), color_mode='grayscale', target_size=(img_rows, img_cols), interpolation='nearest')
    img_array = img_to_array(img, data_format='channels_last', dtype=np.float32)
    images.append(img_array)
    mask = load_img(os.path.join(masksDir, maskName), color_mode='grayscale', target_size=(img_rows, img_cols), interpolation='nearest')
    mask_array = img_to_array(mask, data_format='channels_last', dtype=np.float32)
    masks.append(mask_array)
    
# Reshaping the data
images = np.array(images).reshape(-1, img_rows, img_cols, channels)
masks = np.array(masks).reshape(-1, img_rows, img_cols, channels)

print('Images: {} | Masks: {}'.format(images.shape, masks.shape))

# Save to Numpy Arrays

np.save('./images(8 Bit).npy', images)
np.save('./masks.npy', masks)


    
```
